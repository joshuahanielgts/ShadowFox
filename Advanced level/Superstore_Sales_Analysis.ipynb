# üìä Superstore Sales Data Analysis
This project performs Exploratory Data Analysis (EDA) on a Superstore sales dataset to uncover insights about revenue distribution across products, regions, and customer segments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

## üîç Load the Data

```python
df = pd.read_csv('train.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
df.dropna(how='all', inplace=True)
df.head()
```

## üìã Data Summary

```python
df.info()
df.isnull().sum()
```

## üì¶ Sales by Category

```python
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sales_by_category.plot(kind='bar', color='skyblue', figsize=(8,5))
plt.title('Total Sales by Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

## üìö Sales by Sub-Category

```python
sales_by_subcat = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
sales_by_subcat.plot(kind='bar', color='orange', figsize=(10,5))
plt.title('Total Sales by Sub-Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=90)
plt.show()
```

## üåç Sales by Region

```python
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sales_by_region.plot(kind='bar', color='green', figsize=(6,4))
plt.title('Total Sales by Region')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

## üë• Sales by Segment

```python
sales_by_segment = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
sales_by_segment.plot(kind='bar', color='purple', figsize=(6,4))
plt.title('Total Sales by Segment')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

## ‚úÖ Conclusion
- **Office Supplies** generated the most revenue overall.
- **Phones, Chairs, and Binders** were the top-selling sub-categories.
- **Western and Eastern** regions contributed the most to total sales.
- The **Consumer segment** was the largest revenue source.

These insights can help the business optimize inventory, target high-value segments, and improve marketing strategies.
